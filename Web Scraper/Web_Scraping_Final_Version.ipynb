```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
from time import sleep, time
import lxml
import re
```

```python
search_url = 'https://apps.apple.com/us/genre/ios-games-action/id7001?letter=Q'
data = {'app_name': [],
        'developer': [], 
        'rating': [],
        'rating_count': [],
        'size': [], 
        'age_rating': [], 
        'price':[], 
        'in_app_purchases': [],
        'latest_update': [],
        'description':[],}
```

```python
def export_table_and_print(data):
    table = pd.DataFrame(data, columns=['app_name', 'developer', 'rating', 'rating_count','size', 'price', 'age_rating', 'in_app_purchases', 'latest_update', 'description'])
    table.index = table.index + 1
    table.to_csv('/users/dmauger/Flatiron/Projects/Module5-Project-DM/' +
                 'test3_games.csv', mode='a', encoding='utf-8', index=False)
    print('Scraping done. Here are the results:')
    print(table.info())
    
def game_att(game):

    r = requests.get(game)
    r.encoding = 'utf-8'

    html_content = r.text
    soup = BeautifulSoup(html_content, 'html.parser')

    try:
        name = soup.find('h1', class_='product-header__title app-header__title')
        dev = soup.find('h2', class_="product-header__identity app-header__identity")
        rating = soup.find('div', class_="we-customer-ratings__averages")
        rating_count = soup.find('div', class_="we-customer-ratings__count small-hide medium-show")
        price = soup.find('li', class_="inline-list__item inline-list__item--bulleted app-header__list__item--price")
        purchases = soup.find('li', class_="inline-list__item inline-list__item--bulleted app-header__list__item--in-app-purchase")
        age = [s.string for s in soup.find_all("dd")[5]]
        size = [s.string for s in soup.find_all("dd")[1]]
        description = soup.find('div', class_="section__description")
        latest_update = [a.text.strip() for a in soup.select('div[class$="whats-new__latest"] > div > time')]
    except:
        pass
    
    def cleanhtml(raw_html):
        cleanr = re.compile('<.*?>')
        clean = re.sub(cleanr, '', str(raw_html))
        cleaner = clean.strip()
        cleantext = re.sub('\n','', cleaner)
        return cleantext

    data['app_name'].append(cleanhtml(name))
    data['developer'].append(cleanhtml(dev))
    data['rating'].append(cleanhtml(rating))
    data['rating_count'].append(cleanhtml(rating_count))
    data['price'].append(cleanhtml(price))
    data['in_app_purchases'].append(cleanhtml(purchases))
    data['age_rating'].append(cleanhtml(age))
    data['size'].append(cleanhtml(size))
    data['latest_update'].append(cleanhtml(latest_update))
    data['description'].append(cleanhtml(description))

def parse_page(next_url):
    
    num = 0 
    
    while True:
        page = requests.get(next_url)
        bs = BeautifulSoup(page.content, 'html.parser')     
        test = bs.find('div', class_="grid3-column").find_all('a')
        links = [link.get('href') for link in test]
        
        for game in links:
            game_att(game)

        next_page_text = bs.find('li', class_="paginate-more")

        if next_page_text:
            next_page_url = next_page_text.get('href')
            num += 1
            parse_page(next_page_url)
        else:
            export_table_and_print(data)
            
```
